```python
#features[1 sim 2 não]
#pelo longo?
#perna curta?
#faz auau?

porco1 = [1,1,0]
porco2 = [0,1,1]
porco3 = [0,1,0]

cachorro1 = [0,1,1]
cachorro2 = [1,1,1]
cachorro3 = [0,0,1]


# 1 -> porco  0-> cachorro

treino_x= [porco1, porco2, porco3, cachorro1, cachorro2, cachorro3]
treino_y = [1,1,1,0,0,0] #0 porco 1 cachorro (rótulos)


# quero um estimador que aprenda a classificar (modelo)
from sklearn.svm import LinearSVC


modelo = LinearSVC()
modelo.fit(treino_x, treino_y)
```

```python
#animal misterioso
animal_misterioso = [0,1,1]
modelo.predict([animal_misterioso]) # 1 porco 0 cachorro
```

### A gente precisa de uma nota, igual o professor avaliando uma prova

O misterio1, misterio2 e misterio3 serão utilizados para testes. Nesse caso o modelo não pode conhecê-los

Modelo faz a previsão

```python
misterio1 = [1,1,1] 
misterio2 = [1,1,0] 
misterio3 = [0,1,1] 

teste_x = [misterio1, misterio2, misterio3]
previsoes = modelo.predict(teste_x)
```

```python
# sabia que já era cachorro, porco, porco
teste_y = [0,1,1] #classes corretas - que eu já sei
```

```python
# o que o modelo preveu?
previsoes
```

#### o esperado cachorro , porco, porco


O modelo fez a previsão: cachorro, porco, cachorro (66.6)

# Métricas

```python
from sklearn.metrics import accuracy_score
taxa_de_acerto = accuracy_score(teste_y, previsoes)
```

```python
ta = taxa_de_acerto * 100
print(f"O modelo fez a previsão: cachorro, porco, cachorro ({ta:.2f}%)")
```
